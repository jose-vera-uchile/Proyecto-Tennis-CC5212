```python
import pandas as pd # Importacion estandar de la libreria Pandas
import numpy  as np # Importacion estandar de la libreria NumPy
from google.colab import files


```

#Limpieza de datos#

Limpieza de Matches


```python
url = 'https://media.githubusercontent.com/media/jose-vera-uchile/Proyecto-Tennis-CC5212/main/Raw%20Data/all_matches.csv'
df = pd.read_csv(url)
```

```python
reduced_df=df[['tournament','location','year','player_id', 'opponent_id','sets_won','player_victory','masters']]
reduced_df
```

```python
reduced_df2 = reduced_df.loc[reduced_df['masters']>0]
reduced_df2
```

```python
reduced_df3 = reduced_df2.loc[((reduced_df2['player_victory']=='f')|((reduced_df2['player_victory']=='t')))]
reduced_df3 #Aca tenemos solo las columnas necesarias. y empezamos a quitar nulos. Partimos quitando los partidos que no se sabe quien gano
```

```python
reduced_df3.dropna(how='all') ##Verificamos que no hayan nulos.
```

```python
final_df=reduced_df3.loc[((reduced_df3['sets_won']>=0)&(reduced_df3['sets_won']<=3))] ##Finalmente vemos que no existan Outliers, y que todos los valores de sets ganados sean entre 0 y 3
final_df
```

Con esto tenemos que nuestro dataset de partidos tiene un total de 3.051.885 filas, por lo que luego de la limpieza el total de partidos se mantiene el minimo de columnas necesarias para el proyecto.

Ahora lo exportamos

```python
final_df.to_csv('all_matches_clean.csv', index=False, header=False, encoding = 'utf-8')
files.download('all_matches_clean.csv')
```
